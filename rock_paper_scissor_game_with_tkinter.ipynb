```python
from tkinter import *                       #importing tkinter module
import secrets, random                      #importing secrets module for picking up only one element in an iterable object and random to shuffle it
import win32com.client as wincl              #importing win32 client to use sapi sound (windows pc)


window = Tk()
speak = wincl.Dispatch("SAPI.SpVoice")                        #sapi voice init..
sumofplayer = 0 
sumofpc= 0

selection=["Paper","Scissor","Rock"] 

class autopc:                                        #CLASS FOR PC CHOICE SELECTION
    
    def select(self):
        random.shuffle(selection)
        return secrets.choice(selection)
auto= autopc()

def wish(button): #buttons func
    global sumofpc
    global sumofplayer
    PC= str(auto.select())
    speak.Speak("PC SAY I HAVE"+ PC)                                         #sapi speak returned choice from class pc
    if button =='Rock' and PC=="Scissor":
        sumofplayer= sumofplayer+1
    elif button =="Paper" and PC=="Rock":
        sumofplayer= sumofplayer+1
    elif button=="Scissor" and PC== "Paper":
        sumofplayer=sumofplayer+1
    elif button==PC:
        sumofpc+= 0
    else:
        sumofpc=sumofpc+1
        
    rlabel= Label(window, text= "PLAYER CHOISE IS :  " + button)                       #PRINTING PLAYER_CHOICE
    rlabel.grid(column=1, row=10)
    pclabel= Label(window, text="PC CHOICE IS: " + PC)                                 #PRINTING PC_CHOICE
    pclabel.grid(column=1, row=11)
   
    playerSlabel= Label(window,padx=50 ,pady=10, text="PLAYER WIN:: " + str(sumofplayer))                 #PRINTING PLAYER_SCORES
    playerSlabel.grid(column=1, row=12)
    
    pcSlabel= Label(window,padx=50 ,pady=10, text="PC WIN::  " + str(sumofpc))                                 #PRINTING PC_SCORES
    pcSlabel.grid(column=1, row=13)

    
window.title('Rock_paper_scissor_game_with_tkinter')
window.geometry("400x400")
mylabel = Label(text = "..PLAY WITH YOUR PC...")
mylabel.grid(column=0,row=1)
S1label= Label(window, text=" ")
S1label.grid(column=0, row=2)
mybutton1 = Button(window, text="ROCK",padx=50 ,pady=10 ,bg="red",command=lambda:wish("Rock"))          #PASSING BUTTON NAME TO FUNC WISH
mybutton1.grid(column=1,row=4)

mybutton2 = Button(window, text="SCISSOR",padx=45 ,pady=10, bg= "blue", command=lambda:wish("Scissor"))               #PASSING BUTTON NAME TO FUNC WISH
mybutton2.grid(column=1,row=5)

Slabel= Label(window, text=" ")
Slabel.grid(column=1, row=8)
mybutton3 =Button(window, text="PAPER",padx=50 ,pady=10, bg= "green",command=lambda:wish("Paper"))                      #PASSING BUTTON NAME TO FUNC WISH
mybutton3.grid(column=1,row=6)





window.mainloop()
```
